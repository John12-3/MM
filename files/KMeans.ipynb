```python
def getsquaredistance(point1,point2):
  return round(((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2),4)
```

```python
def getdistancefrompoints(centroids,datapoints):
  distancefromcluster = []
  intermediate_result = []

  for i in centroids:
    for j in datapoints:
      intermediate_result.append(getsquaredistance(i,j))
    distancefromcluster.append(intermediate_result)
    intermediate_result = []
  return distancefromcluster
```

```python
def printResults(centroids,point_to_cluster_mapping):
  for i in range(len(centroids)):
    print("centroids ",i,centroids[i])
  for i in point_to_cluster_mapping:
    print("Points : ",i," Cluster ",point_to_cluster_mapping[i])
```

```python
def kmeansclustering(centroids,datapoints):
  '''
  Driver code for K means clusering
  '''
  distancefromcluster = getdistancefrompoints(centroids,datapoints)
  point_to_cluster_mapping = {}
  max_valued_cluster = 0
  for i in range(len(datapoints)):
    point_to_cluster_mapping[i] = None
    for cluster in range(len(centroids)):
      if distancefromcluster[cluster][i] < distancefromcluster[max_valued_cluster][i]:
        max_valued_cluster = cluster
    point_to_cluster_mapping[i] = max_valued_cluster

  # compute the new centroids by average of 2 points
  cluster_counter = 0
  for i in range(len(centroids)):
    for j in point_to_cluster_mapping:
      if point_to_cluster_mapping[j] == i:
        centroids[i][0] += datapoints[i][0]
        centroids[i][1] += datapoints[i][1]
        cluster_counter +=1
    if cluster_counter != 0:
      centroids[i][0] = round(centroids[i][0]/cluster_counter,4)
      centroids[i][1] = round(centroids[i][1]/cluster_counter,4)
      cluster_counter = 0
  
  printResults(centroids,point_to_cluster_mapping)
  return centroids
```

```python
def kmeaniterator(centroids,datapoints):
  old_centroids = centroids
  new_centroids = centroids
  iterator = 0

  while iterator != 15:
    iterator += 1
    print("Iterator " , iterator)
    old_centroids = new_centroids
    new_centroids = kmeansclustering(new_centroids,datapoints)
```

```python
centroids = [[6.2,3.2],[6.6,3.7],[6.5,3.0]]
datapoints = [[5.9,3.2],[4.6,2.9],[6.2,2.8],[4.7,3.2],[5.5,4.2],[5.0,3.0],[4.9,3.1],[6.7,3.1],[5.1,3.8],[6.0,3.0]]
kmeaniterator(centroids,datapoints)
```
