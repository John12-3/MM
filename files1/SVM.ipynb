```python
import numpy as np
import pandas as pd
```

```python
#  we need to have 3 types of kernals
df = pd.read_csv("/content/sample_data/SVM_Dataset_Glass.csv")
df.head()
```

```python
# now take the x and y
x = df.drop(["Type"],axis=1)
x
```

```python
y = df["Type"]
y
```

```python
# create test , train and split the datase
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn import svm
from sklearn.svm import SVC
```

```python
model1 = SVC(kernel="linear")
model1.fit(xtrain,ytrain)
y_pred = model1.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print(accuracy_score(ytest,y_pred))
print(confusion_matrix(ytest,y_pred))
print(classification_report(ytest,y_pred))
```

```python
# Creating the models for kernels like rbf , poly and sigmoid
model2 = SVC(kernel='sigmoid')
model3 = SVC(kernel='rbf')
model4 = SVC(kernel="poly")
```

```python
model2.fit(xtrain,ytrain)
model3.fit(xtrain,ytrain)
model4.fit(xtrain,ytrain)
```

```python
y_pred2 = model2.predict(xtest)
y_pred3 = model2.predict(xtest)
y_pred4 = model2.predict(xtest)
```

```python
print("Prediction of model2 is " , accuracy_score(ytest,y_pred2))
print("Prediction of model2 is " , accuracy_score(ytest,y_pred3))
print("Prediction of model2 is " , accuracy_score(ytest,y_pred4))
```
